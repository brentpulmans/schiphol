```python
import pandas as pd
import numpy as np

path = 'C:/Users/bmpka/Documents/Thomas More/3 BIT/Big Data/Schiphol/flight_subset.csv'

f = open(path, 'r')

data = f.read()

rows = data.split('\n')

print(rows[0:10])
```

```python
vluchtnummers = []

for row in rows:    
    vluchtparameters = row.split(',')
    vluchtnummers.append(vluchtparameters[10:11])
    
print(vluchtnummers[0:10])
```

```python
aantalVluchtnummers = len(vluchtnummers)

print(aantalVluchtnummers)
```

```python
aantalUniekeVluchtnummers = len(np.unique(vluchtnummers))

print(aantalUniekeVluchtnummers)
```

```python
datum = '2017-10-01'

# alle vluchten voor hierboven bepaalde datum ophalen
vluchtenVoorDatum = []

for row in rows:    
    vluchtparameters = row.split(',')
    
    # Hieronder in if: nakijken of opgegeven datum voorkomt in veld scheduleDate
    if datum in vluchtparameters[3:4]:
        #print(vluchtparameters)
        vluchtenVoorDatum.append(row)
        
print(vluchtenVoorDatum)
```

```python
print("Aantal vluchtnummers voor " + datum + ": " + str(len(vluchtenVoorDatum)))
```

```python
vluchtnummersVoorDatum = []

for vlucht in vluchtenVoorDatum:
    vluchtparameters = vlucht.split(',')
    vluchtnummersVoorDatum.append(vluchtparameters[10:11])

print("Aantal unieke vluchten voor " + datum + ": " + str(len(np.unique(vluchtnummersVoorDatum))))
```

```python
# Info over codeshares op opgegeven datum

print("Aantal vluchten als codeshare (geen echte vlucht) op " + datum + ": " + str(len(vluchtenVoorDatum) - len(np.unique(vluchtnummersVoorDatum))))
print("Aantal vluchten als codeshare (percentueel) op " + datum + ": " + str((len(vluchtenVoorDatum) - len(np.unique(vluchtnummersVoorDatum))) / len(vluchtenVoorDatum) * 100) + "%")
```

```python
# Unieke terminals bepalen
terminalsFromFile = []

#vluchtenPerTerminal = sorted(vluchtenVoorDatum, key=lambda vlucht: vlucht[15])
for row in vluchtenVoorDatum:
    parameters = row.split(',')
    terminalsFromFile.append(parameters[15:16])

terminals = np.unique(terminalsFromFile)
print(terminals)
```

```python
terminal1 = 0
terminal2 = 0
terminal3 = 0
terminal4 = 0
terminalNULL = 0

# Functies opstellen voor tellen vluchten per terminal
# In iedere functie wordt eerst aangegeven met global welke variabele gebruikt gaat worden,
# anders denkt Python dat het een lokale variabele zal behandelen en wordt een error gegeven
def Terminal1():
    global terminal1
    terminal1 += 1

def Terminal2():
    global terminal2
    terminal2 += 1

def Terminal3():
    global terminal3
    terminal3 += 1

def Terminal4():
    global terminal4
    terminal4 += 1
    
def TerminalOnbekend():
    global terminalNULL
    terminalNULL += 1

# map the inputs to the function blocks
options = {1 : Terminal1,
           2 : Terminal2,
           3 : Terminal3,
           4 : Terminal4,
           0 : TerminalOnbekend
}
```

```python
# Aantal vertrekken vanuit iedere terminal
for row in vluchtenVoorDatum:
    parameters = row.split(',')
    if(parameters[15:16][0] == 'NULL'):
        options[0]()
    else:
        options[int(parameters[15:16][0])]()
```

```python
print("Aantal vluchten in terminal 1: " + str(terminal1))
print("Aantal vluchten in terminal 2: " + str(terminal2))
print("Aantal vluchten in terminal 3: " + str(terminal3))
print("Aantal vluchten in terminal 4: " + str(terminal4))
print("Aantal vluchten zonder terminal (regering, militair ...): " + str(terminalNULL))
```
